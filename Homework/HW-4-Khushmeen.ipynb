For these simulations I analyzed the `ener.edr` file using the command `gmx energy -f ener.edr` 

However I recorded the same output for PE,KE and TE for each of the 3 `energy.xvg` files.

```python
data= pd.read_table('energy.xvg',delim_whitespace=True,skiprows=25,names=['Time','PE','KE','TE'])
data
```

The plots for the various energy values are shown.

```python
fig = plt.figure()
fig.set_figwidth(18)
fig.set_figheight(18)
plt.rcParams.update({'font.size': 11})

axes = fig.add_subplot(221)
axes.plot(data.Time,data.PE)
axes.set_xlabel('TIME')
axes.set_ylabel('PE')
axes.set_title('2')
axes.grid()
axes = fig.add_subplot(222)
axes.plot(data.Time,data.KE)
axes.set_xlabel('TIME')
axes.set_ylabel('KE')
axes.set_title('3')
axes.grid()
axes = fig.add_subplot(223)
axes.plot(data.Time,data.TE)
axes.set_xlabel('TIME')
axes.set_ylabel('TE')
axes.set_title('4')
axes.grid()
fig.subplots_adjust(hspace=0.5)
fig.subplots_adjust(wspace=0.5)
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(8)
fig.set_figheight(8)
plt.rcParams.update({'font.size': 11})

plt.plot(data['Time'],data['PE'],label='Ep')
plt.plot(data['Time'],data['KE'],label='Ek')
plt.plot(data['Time'],data['TE'],label='Etot')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Potential, Kinetic and Total Energy')
plt.legend()
```
