```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Form a hypothesis on how this analysis will be different for the three runs and justify (1 paragraph)

### Independent variable: pressure (1, 100, 500 bar)
### Dependent variable: volume

It is expected that as the pressure increases, the volume will decrease. Temperature and number of molecules remain constant across all simulations; therefore, to couple the system's pressure using a barostat, the average volume for each simulation should decrease as pressure increases.

## Plots:

```python
volume_1 = pd.read_csv('hw4/volume_1.xvg',skiprows = 23,skipinitialspace=True,\
                   delimiter=' ',usecols=[0,1],header=None,names=["time","volume"])
volume_100 = pd.read_csv('hw4/volume_100.xvg',skiprows = 23,skipinitialspace=True,\
                   delimiter=' ',usecols=[0,1],header=None,names=["time","volume"])
volume_500 = pd.read_csv('hw4/volume_500.xvg',skiprows = 23,skipinitialspace=True,\
                   delimiter=' ',usecols=[0,1],header=None,names=["time","volume"])

pressure_1 = pd.read_csv('hw4/pressure_1.xvg',skiprows = 23,skipinitialspace=True,\
                   delimiter=' ',usecols=[0,1],header=None,names=["time","pressure"])
pressure_100 = pd.read_csv('hw4/pressure_100.xvg',skiprows = 23,skipinitialspace=True,\
                   delimiter=' ',usecols=[0,1],header=None,names=["time","pressure"])
pressure_500 = pd.read_csv('hw4/pressure_500.xvg',skiprows = 23,skipinitialspace=True,\
                   delimiter=' ',usecols=[0,1],header=None,names=["time","pressure"])


plt.figure(figsize=(10,5))
plt.subplot(231)
plt.plot(volume_1.time,volume_1.volume)
plt.ylabel('$\mathrm{Volume (nm^3)}$',fontsize=15,position=(1,0.5))
plt.ylim(26,30)

plt.subplot(232)
plt.plot(volume_100.time,volume_100.volume)
plt.ylim(26,30)

plt.subplot(233)
plt.plot(volume_500.time,volume_500.volume)
plt.ylim(26,30)

plt.subplot(234)
plt.plot(pressure_1.time,pressure_1.pressure)
plt.ylabel('$\mathrm{Pressure (bar)}$',fontsize=15,position=(-10,0.5))
plt.ylim(-1500,2500)

plt.subplot(235)
plt.plot(pressure_100.time,pressure_100.pressure)
plt.ylim(-1500,2500)
plt.xlabel('$\mathrm{Time (ps)}$',fontsize=15,position=(0.5,1))

plt.subplot(236)
plt.plot(pressure_500.time,pressure_500.pressure)
plt.ylim(-1500,2500)

plt.tight_layout()
```
