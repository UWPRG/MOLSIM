```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('energies.csv', skiprows=27)
#### Time, Potential Energy, Kinetic Energy, Total Energy, Pressure, Density
time=data[:,0]
PE=data[:,1]
KE=data[:,2]
totalE=data[:,3]
Pressure=data[:,4]
density=data[:,5]
#total drift for this simulation = -0.00555237 kJ/mol
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,PE,time,KE,time,totalE,'r-')
plt.rc('font',size=14)
plt.title('Energies vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Energy (KJ/mol)')
plt.legend(['Potential Energy','Kinetic Energy','Total Energy'],loc='best')
plt.grid()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,Pressure,'r-')
plt.rc('font',size=14)
plt.title('Pressure vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (KJ/mol)')
plt.grid()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,density,'r-')
plt.rc('font',size=14)
plt.title('Density vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg/m^3)')
plt.grid()
```
