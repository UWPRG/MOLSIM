```python
dt_energy_drift = np.genfromtxt('dt_energydrift.csv')
temp_energy_drift = np.genfromtxt('temp_energydrift.csv')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(dt_energy_drift[:,0],dt_energy_drift[:,1],'*',markersize='10')
plt.title('Energy Drift as dt changes')
plt.ylabel('Energy Drift')
plt.xlabel('time step - dt (ps)')
plt.ylim(-0.6,0.1)

plt.subplot(122)
plt.plot(temp_energy_drift[:,0],temp_energy_drift[:,1],'*',markersize='10')
plt.title('Energy Drift as temp changes')
plt.xlabel('Temperature (K)');
```

# Energy Drift
## dt changes
Energy drift becomes more and more negative as dt increases because there is a greater and greater time gap between force and energy calculations, allowing atoms to drift more closely together than when a much smaller time step is used.

## Temperature changes
As temperature is increased, the system starts with greater kinetic energy causing higher atomic velocities. As those velocities get greater and greater, a similar effect is seen as before with changes in dt. The time step becomes too big and atoms drift closer than normal, thereby causing greater forces than usual.
