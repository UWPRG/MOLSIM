```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt("energy.csv")
data_2 = np.loadtxt("energy_dt.csv")
```

```python
print (np.shape(data), np.shape(data_2))
```

```python
##add units

#NVE System
#[row, column]
time=data[:,0]
LJ=data[:,1]
Potential = data[:,2]
Kinetic = data[:,3]
Total= data[:,4]
Temp=data[:,5]

##broken system
time_2=data_2[:,0]
Potential_2 = data_2[:,2]
Kinetic_2 = data_2[:,3]
Total_2= data_2[:,4]
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,LJ)
plt.title('Lennard Jones')
plt.xlabel('time (ps)')
plt.ylabel('LJ')
plt.grid()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,Potential, label = 'Potential Energy')
plt.plot(time,Kinetic, label = 'Kinetic Energy')
plt.plot(time, Total, label = 'Total Energy')
plt.legend()
plt.title('Energy')
plt.xlabel('time (ps)')
plt.ylabel('Energy (kJ/mol)')
plt.grid()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,Temp)
plt.title('Temperature')
plt.xlabel('time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
```

```python
##with an increase in the timestep 

plt.figure(figsize=(6, 6))
plt.plot(time_2,Potential_2, label = 'Potential Energy')
plt.plot(time_2,Kinetic_2, label = 'Kinetic Energy')
plt.plot(time_2, Total_2, label = 'Total Energy')
plt.legend()
plt.title('Energy')
plt.xlabel('time (ps)')
plt.ylabel('Energy (kJ/mol)')
plt.grid()
```

```python
data_bad = pd.read_csv('corruptdata.csv', delimiter = " ", names = ["timestep", "PE", "KE", "Total"])
data_bad
```

```python
##Energy drift as a function of the timestep 

plt.figure(figsize=(6, 6))
plt.scatter(data_bad.timestep,data_bad.PE, label = 'Potential Energy', c='b')
plt.scatter(data_bad.timestep,data_bad.KE, label = 'Kinetic Energy', c='r')
plt.scatter(data_bad.timestep, data_bad.Total, label = 'Total Energy', c='g')
plt.legend(scatterpoints = 1, loc= 'upper right', fontsize='8')
plt.title('Energy Drift')
plt.xlabel('time step(ps)')
plt.ylabel('Energy Drift (kJ/mol)')
plt.grid()
plt.ylim([-6,1])
plt.xlim([0, 0.06])
```
