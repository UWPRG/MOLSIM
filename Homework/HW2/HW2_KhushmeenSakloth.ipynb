```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
data= pd.read_table('energy.xvg',delim_whitespace=True,skiprows=27,names=['Time','LJ','Potential','KE','TE','Temperature'])
```

```python
data

```

```python
fig = plt.figure()
fig.set_figwidth(13)
fig.set_figheight(13)
plt.rcParams.update({'font.size': 17})
axes = fig.add_subplot(221)
axes.plot(data.Time,data.LJ)
axes.set_xlabel('TIME')
axes.set_ylabel('LJ')
axes.set_title('1')
axes.grid()
axes = fig.add_subplot(222)
axes.plot(data.Time,data.Potential)
axes.set_xlabel('TIME')
axes.set_ylabel('POTENTIAL')
axes.set_title('2')
axes.grid()
axes = fig.add_subplot(223)
axes.plot(data.Time,data.KE)
axes.set_xlabel('TIME')
axes.set_ylabel('KE')
axes.set_title('3')
axes.grid()
axes = fig.add_subplot(224)
axes.plot(data.Time,data.TE)
axes.set_xlabel('TIME')
axes.set_ylabel('TE')
axes.set_title('4')
axes.grid()
fig.subplots_adjust(hspace=0.5)
fig.subplots_adjust(wspace=0.5)
plt.show()
```

```python
plt.plot(data['Time'],data['Potential'],label='Ep')
plt.plot(data['Time'],data['KE'],label='Ek')
plt.plot(data['Time'],data['TE'],label='Etot')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Potential, Kinetic and Total Energy')
plt.legend()
```

Energy Drift v/s Timestep


```python
e_drifts=[-0.00555237,-0.547443,-0.115457,-0.00967618,-0.00531041]
dt=[0.0005,0.05,0.01,0.001,0.0001]
plt.figure(figsize=(8,8))
plt.scatter(dt,e_drifts)
plt.xlabel('dt')
plt.ylabel('E_drifts')
plt.title('Energy Drift vs. Time Step')
plt.grid()
```

By increasing the time step , the energy drift values become more negative. Also the time taken to compute the simulations increases with increasing time step. (The computer did not respond for a time step=10).By varying temperature too the energy drift changes.
