```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('energy2.csv')#, delimiter=' ')
#### Time, LJ, Potential Energy, Kinetic Energy, Total Energy, Temperature
time=data[:,0]
LJ=data[:,1]
PE=data[:,2]
KE=data[:,3]
totalE=data[:,4]
temp=data[:,5]
#total drift for this simulation = -0.00555237 kJ/mol
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,LJ,'r-')
plt.rc('font',size=14)
plt.title('Lennard Jones Potential vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Lennard Jones Potential (KJ/mol)')
plt.grid()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,PE,'r-',time,KE,'g-',time,totalE,'b-')
plt.rc('font',size=14)
plt.title('Energies vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy(kJ/mol)')
plt.grid()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(time,temp,'r-')
plt.rc('font',size=14)
plt.title('Temperature vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Temp (K)')
plt.grid()
```

```python
dts=[0.0005,0.01,0.0005,0.05,0.001,0.02,0.04,0.03,0.005]
drifts=[-0.00555237,-1.72393,-0.0156489,-5.63324,-0.0156489,-3.97954,-5.49672,-4.99592,-0.502358]

plt.figure(figsize=(6, 6))
plt.scatter(dts,drifts)
plt.rc('font',size=14)
plt.title('Energies vs. Time Steps')
plt.xlabel('Time steps (ps)')
plt.ylabel('Total Energy drifts (kJ/mol)')
plt.grid()
```

Changing dt to large time steps and increasing temperature to extremes will both lead to large energy drifts. However, the simulations took too long to run for the large temperatures. These result in drifts because the large time step makes the simulation run longer and more drift will occur. Extreme temperatures also leads to drift because of the excessive amount of energy and large variations in the energies will cause the drift.
