## Using Plumed , the evolution of the phi_angle with time is shown

```python
data= pd.read_table('Khushmeen',delim_whitespace=True,skiprows=3,names=['TIME','PHI'])
```

```python
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(20)
plt.rcParams.update({'font.size': 17})
axes = fig.add_subplot(221)
axes.plot(data.TIME,data.PHI)
axes.set_xlabel('TIME')
axes.set_ylabel('PHI')
axes.set_title('1')
axes.grid()
```

I read the paper 'Efficient Sampling of High-Dimensional Free-Energy Landscapes with Parallel Bias Metadynamics' and the following are the answers to the 3 questions.

  1)
  * In standard Metadynamics(MetaD) an external bias potential is introduced to accelerate sampling along collective variables (CV) while in parallel bias methods (PBMetaD), simultaneous multiple low dimensional bias potentials are applied.
  * In standard MetaD, the bias potential is constructed in the space of a few CVs as a sum of Gaussians deposited along the system trajectory. However in PBMetaD, there will be more than one bias potential and the sum of each ones Gaussians along the trajectory is required.
  

2) Advantages of PBMetaD
* It helps to cross high barriers in energy landscapes in a time scale of general simulations.  
* The free energies can be calculated from the bias potentials and the full high dimensional free energy can be calculated using standard reweighting techiniques.
* It is simple and flexible to use.  

  Disadvantages of PBMetaD
* Perhaps a multidimensional gaussian of 2 or 3 CVs may converge faster in MetaD
