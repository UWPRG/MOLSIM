````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
plumed=pd.read_csv('PLUMED_PHI',delimiter=' ',skiprows=3,names=["time","plumed_phi","calc_phi"],usecols=[1,2,3])
gmx=pd.read_csv('phiALA2.xvg',skiprows=16,names=["time","phi"],delim_whitespace=True)
vmd=pd.read_csv('vmd_phi.dat',delim_whitespace=True,names=["frame","phi"])
````

````python
plt.figure(figsize=(8,6))

plt.subplot(221)
plt.plot(plumed.time,plumed.plumed_phi)
plt.ylabel('$\mathrm{\phi \ (radians)}$',fontsize=14)
plt.title('Plumed')

plt.subplot(222)
plt.plot(plumed.time,plumed.calc_phi)
plt.title('Plumed using MATHEVAL')

plt.subplot(223)
plt.plot(gmx.time,gmx.phi)
plt.title('GROMACS')
plt.ylabel('$\mathrm{\phi \ (radians)}$',fontsize=14)
plt.xlabel('time (ps)',fontsize=12)

plt.subplot(224)
plt.plot((vmd.frame)/10,np.radians(vmd.phi))
plt.title('VMD')
plt.xlabel('time (ps)',fontsize=12)
plt.xlim(0,1000)

plt.tight_layout();
````

````python
plt.figure(figsize=(8,6))
plt.plot(plumed.time,plumed.plumed_phi,label='Plumed')
plt.plot(plumed.time,plumed.calc_phi,label='Plumed + MATHEVAL')
plt.plot(gmx.time,gmx.phi,label='GROMACS')
plt.plot((vmd.frame)/10,np.radians(vmd.phi),label='VMD')
plt.xlim(0,1000)
plt.legend(bbox_to_anchor=(1.04,-0.11),ncol=4)
plt.xlabel('time (ps)',fontsize=12)
plt.ylabel('$\mathrm{\phi \ (radians)}$',fontsize=15)
plt.title('All Plots',fontsize=20);
````

## Plumed $\phi$ calculation format:

```
FUNC=(atan(y/x)*step(x))
     +((atan(y/x)+3.14159265)*step(y)*step(-x))
     +((atan(y/x)-3.14159265)*step(-y)*step(-x))
```

# Part 2 - Well-Tempered Metadynamics

### 1. How does WT-MetaD  differ from MetaD?
- WT-MetaD varies the gaussian hill height as a function of time

### 2. What are the advantages/disadvantages compared to vanilla MetaD?
#### Advantages
- Converges to the exact free energy surface in the limit of long simulation time
- Greater control over visiting physically relevent phase space
- Converges faster due to a flat bias potential when hill heights approach zero, allowing for a diffusive state

#### Disadvantages
- You need to have a good understanding of your system in order to converge
