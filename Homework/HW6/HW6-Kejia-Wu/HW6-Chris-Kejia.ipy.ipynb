```python
vmd = np.loadtxt('di_vmd.txt')
frame = vmd[:,0]
degree = vmd[:,1]
phi = [math.radians(d) for d in degree]
time = frame / 10

plt.figure(figsize=(8, 8))
plt.plot(time, phi , 'r.')
plt.rc('font', size=10)
plt.title('Dihedral Angle from VMD vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (rad)')
plt.grid()
```
