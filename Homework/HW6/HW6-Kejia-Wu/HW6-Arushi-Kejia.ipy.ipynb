We need to find out:
* The purpose of each file (aminoacids.rtp , forcefield.itp, …)
* Content of each file, and its format ( File X has a list of atom names, atomic weights …)
* Units used in each file ( kcal/mol or kJ/mol, nm or Angstrom, … )

## Solutions:
* aminoacids.arn  
* dna.arn  
* rna.arn  

Force fields often use atom names that do not follow IUPAC or PDB convention. The .arn database is used to translate the atom names in the coordinate file to the force field names. Atoms that are not listed keep their names. The file has three columns: the building block name, the old atom name, and the new atom name, respectively.  


* aminoacids.c.tdb  

The termini databases are stored in aminoacids.n.tdb and aminoacids.c.tdb for the N and C termini respectively. They contain information for the pdb2gmx program on how to connect new
atoms to existing ones, which atoms should be removed or changed, and which bonded interactions should be added.  


* aminoacids.n.tdb  

For N termini. They contain information on how to connect new atoms to exisitng ones, which atoms should be removed or changdd, and which bonded interactions should be added. The file is organized in blocks, each with a header specifying the name of the block. These blocks correspond to different types of termini that can be added to a molecule. In this example [ COO- ] is the first block, corresponding to changing the terminal carbon atom into a deprotonated carboxyl group. [ None ] is the second terminus type, corresponding to a terminus that leaves the molecule as it is.  

[ replace ] : replace an existing atom by one with a different atom type, atom name, charge, and or mass. For each atom to be replaced on line should be entered with the following fields: 
    - name of the atom to be replaced
    - new atom name
    - new atom type
    - new mass
    - new charge

[ add ] : add new atoms. For each added atom. a two line entry is necessary. The first line contains the same fields as an entry in the hydrogen database ( name of the new atom, number of atoms, type of addition, control atoms) but the possible types of addition are extended by two more, specifically for C-termini additions: see manual page 148. 
After this line, another line follwos that specifies the details of the added atoms, in the same way as for replaciing atoms
    - atom type
    - atom mass
    - charge
    - charge group (optional)
Like in the hydrogen database, when more than one atom is connected to an existing one, a number will be appended to the end of the atom name. When the chare group field is left out, the added atom will have the same charge group number as the atom that it is bonded to.

[ delete ] : delete existing atoms. one atom name per line.

[ bonds ] [ angles ] [ dihedrals ] [ imporpers ] : add additional bonded parameters. The format is identical to that used in the .rtp file.

* aminoacids.hdb
* dna.hdb
* rna.hdb   

The hydrogen database is stored in .hdb files. It contains information for the pdb2gmx program on how to connect hydrogen atoms to existing atoms. The H atom has to be listed explicitly since the behavior is protein specific and not general. If more than one hydrogen atom is connected to the same atom, a number will be added to the end of the hydrogen atom name. For example adding two hydrogen atoms to ND2, the hydrogen atoms will be named HD21 and HD22.  

On the first line, we see the residue name (ALA or ARG) and the number of kinds of hydrogen atoms that may be added to this residue by the database. After that follows one line for each addition on which we see:
    - the number of H atoms added
    - the method for adding H atoms (see page 145 of manaul for methods)
    - the name of the new H atom
    - Three or four control atoms ( i,j,k,l) where the first always is the atom to which the H atoms are connected. The other two or three depend on the code selected. For water, there is only one control atom.  
    

* aminoacids.r2b
* dna.r2b
* rna.r2b  

Each force field has its own naming convention for residues. Most residues have consistent
naming, but some, especially those with different protonation states, can have many different
names. The .r2b files are used to convert standard residue names to the force field build block names. The .r2b can contain 2 or 5 columns. The 2 column format has the residue name in the first column and the building block in the second. The 5 column format has 3 additional columns with the building block for the residue occuring in the N-terminus, C-terminus and both termini and the same time (single residue molecule). If one or more of the terminal versions are not present, a dash is entered. If no .r2b is present in the force field directory or a residue is not listed, the building block name is assumed to be identical to the residue name.  


* aminoacids.rtp
* dna.rtp
* rna.rtp  

The files holding the residue databases have the extension .rtp. So the residue database file
contains information (bonds, charges, charge groups, and improper dihedrals) for a frequently used building block.   


* aminoacids.vsd  

Decides the geometries and parameters with which to add virtual site hydrogens. For more complex virtual site constructs, we also need information about the equilibrium bond lengths and angles for all atoms in the side chain. The first couple of sections specify which mass centers (typically called MCH3/MNH3) to use for CH3, NH3 and NH2 groups. We do not need to specify the actual parameters, just the type of mass center to use. To accomplish this, there are three section names [ CH3 ], [ NH3 ], and [ NH2 ]. For each of these we expect three columns. The first column is the atom type bound to the 2/3 hydrogens, the second column is the next heavy atom type which this is bound, and the third column is the type of mass center to use.   

The second part of the virtual site database contains explicit equilibrium bond lengths and angles for pairs/triplets of atoms in aromatic side chains.These sections are named after the short aminoacid names ([ PHE ], [ TYR ], [ TRP ], [ HID ], [ HIE ], [ HIP ]), and simplycontain 2 or 3 columns with atom names, followed by a number specifying the bond length (in nm) or angle (in degrees).  


* atomtypes.atp  

An atom type file (extension .atp) from the force field directory. Each force field defines a set of atom types, which have a characteristic name or number, and mass (in a.m.u.). These listings are found in the atomtypes.atp file (.atp = atom type parameter file).   


* ffbonded.itp  

The bonded parameters (i.e. bonds, bond angles, improper and proper dihedrals) are listed in
ffbonded.itp. The entries in this database describe, respectively, the atom types in the interactions, the type of the interaction, and the parameters associated with that interaction. These parameters are then read by grompp when processing a topology and applied to the relevant bonded parameters, i.e. bondtypes are applied to entries in the [ bonds ] directive, etc. Any bonded parameter that is missing from the relevant [ *type ] directive generates a fatal error. More specifically, the first and second columns name the atoms that are bonded. The third column defines the bonded function type. The fourth column is the equilibrium bond distance (nm). the fifth column is the bond spring constant (kJ/mol/nm^2).  


* ffnonbonded.itp  

The non-bonded parameters consist of the van derWaals parameters V (c6 or σ, depending on the
combination rule) and W (c12 or ε), as listed in the file ffnonbonded.itp. First column is the name of the atom. Second column is the atmoic number of the atom. Third Column is mass of the atom in amu. Fourth Column is the charge of the atom. Fifth Column is the particle type (atoms=A, shells=S, virtual interacting sites=V or D (such as benzene rings)). Sixth column is the VdW sigma (in nm). Seventh column is the VdW epsilon (in nm).  


* forcefield.doc  

The file {\tt forcefield.doc} may be present, and if so, its first line will be used by {\tt pdb2gmx} to present a short description to the user to help in choosing a force field. Otherwise, the user can choose a force field with the {\tt -ff xxx} command-line argument to {\tt pdb2gmx}, which indicates that a force field in a {\tt xxx.ff} directory is desired. {\tt pdb2gmx} will search first in the working directory, then in the {\gromacs} {\tt share/top} directory, and use the first matching {\tt xxx.ff} directory found.  


* forcefield.itp  

The GROMACS program pdb2gmx generates a topology for the input coordinate file. Several
formats are supported for that coordinate file, but *.pdb is the most commonly-used format (hence the name pdb2gmx). pdb2gmx searches for force fields in subdirectories of the GROMACS
share/top directory and your working directory. Force fields are recognized from the file
forcefield.itp in a directory with the extension .ff. Two general files are read by pdb2gmx: an atom type file (extension .atp) from the force field directory, and a file called residuetypes.dat from either the working directory, or the GROMACS share/top directory. residuetypes.dat determines which residue names are considered protein, DNA, RNA, water, and ions.  


* gbsa.itp  

For implicit solvent parameters. the atom type is listed in the first column. Values in columns 2-4 are not currently used. They pertain to more elaborate surface area algorithms. Column 5 contains the atomic van der Waals radii, which are used in computing the Born radii. The dielectric offset is specified in the .mdp file, and gets subtracted from the input van der Waals radii for the different Born radii methods. Column 6 is the scale factor for the HCT and OBC models. The values are taken from the Tinker implementation of the HCT pairwise scaling method. This method has been modified such that the scaling factors have been adjusted to minimize differences between analytical surface areas and GB using the HCT algorithm.  
 
 
* ions.itp
* spc.itp
* spce.itp
* tip3p.itp
* tip4p.itp
* urea.itp
* tip4pew.itp
* tip5p.itp  

[ moleculetype ] : defines the name of your molecule in this file. (SOL). nrexcl = 2 stands for excluding non-bonded interactions between atomes that are no further than 2 bonds away.  

[ atoms ] : defines the molecule. id assigns a number to each atom. at type names each atom. res nr is the residue number of the molecule. res name is the residue name (in this case, a solvent). at name is the unique atom name for each of the atoms in the molecule. cg nr is the charge group for each of the atoms (each charge group must add to 0 for neutral molecules). Charge is the actual charge of the atoms. mass is the atomic mass of the atom. 

[ settles ] : first defines the first atom of the water molecule. The settle funct is always 1, and the distance between O-H and H-H distances must be given. [nm]

[ exlusions ] : When more interactions need to be excluded, or some pair interactions should not be generates, an [ exclusions ] field can be added, followed by pairs of atom names on separate lines. All non-bonded and pair interactions between these atoms will be excluded.

[ bonds ] : i is the id number of the first molecule in the bond. j is the id number of the second molecule in the bond. funct is the type of bond it is (single, double, etc). Length is the equilibrium length of bond in nm. force constant is the spring constant of the bond in kJ/mol/nm^2. 

[ angles ] : i is the one of the end atoms, j is the atom in the center, and k is the other end atom (by id number). the length is the equilibrium bond length. The force constant is the spring constant of bending the 3 atoms.
