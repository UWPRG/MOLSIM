#### aminoacids.arn
1.Atom renaming database, translate atom names in coordinate file to forcfield names <br />
2.[columns] Building block name (residue), old atom name(gromacs), new atom name (forcefield) <br />
3.N/A<br />

#### aminoacids.c.tbd
1.C-Termini database, parameters that describe how to connect new atoms to existing ones <br />
2.[blocks] header(different termini that can be added), terminal action(replace, add, etc.), detail of added atoms (atom type, mass, charge, charge group), <br />
3.mass = amu<br />

#### aminoacids.hdb
1.Hydrogen database, contains info on how to connect H atoms to existing atoms<br />
2.[columns] Residue name, number of H atoms that can be added, number of H atoms added, method for adding H atoms, name of new H atom, control atom (3 or 4, i,j,k,l)<br />
3.N/A<br />

#### aminoacids.n.tbd
1.N-Termini database, same as C termini database but for C-termnini<br />
2.See "aminoacids.c.tbd"<br />
3.See "aminoacids.c.tbd"<br />

#### aminoacids.r2b
1.Used to convert standard residue names to force-field block names <br />
2.[columns] old residue name (gromacs), building block name, n-term name, c-term name<br />
3.N/A<br />
#### aminoacids.rtp
1.Holds residue databases, information for frequently used building blocks (bonds, charges, charge groups, improper dihedrals)<br />
2.[columns and blocks] (freeformat, only restriction is there must be more than one line)<br />
- [columns] default bondtype, charge, charge group, default angletype, default proper dihedral, default improper dihedral, generation of dihedrals from bonding, number of neighbors to exclude from nonbonded interactions, bonded interactions, dihedral interactions
- [blocks] different residues <br />

3.angles = radians<br />

#### aminoacids.vsd

1.Virtual site database, decides geometries and parameters with which to add virtual site hydrogens<br />
2.[columns and blocks]<br />
- [blocks] section names for CH3, NH3, NH2 - each has the following three columns)
- [columns] atom type bound to the H's, next heavy atom type which is bound, center of mass to use <br />
NH2 has the option for the 2nd column to specify planar, which will not use a mass center). The second part of the virtual site database contains explicit equilibrium bond lengths and angles for pairs/triplets of atoms in aromatic side chains <br />
3.bond length = nm or angle = degrees.<br />

#### atomtypes.atp
1.Atom type parameter, defines a set of atom types for each forcefield <br />
2.[column] name, mass <br />
3.mass =amu <br />

#### gbsa.itp
1.Do a simulation with implicit solvent using the Generalized Born formalism. Three different methods for calculating the Born radii are available, Still, HCT and OBC. These are specified with the gb-algorithm field. The non-polar solvation is specified with the sa-algorithm field. <br /> 
2.[column] atom type, (Values in columns 1-3 are not currently used. They pertain to more elaborate surface area algorithms,the one from Qiu et al.), atomic Van Der Waals radii, scale factor for HTC AND OBC models <br />
3.nm

#### tip3p.itp
1.Topolgy file for the tip3p water model<br />
2.[blocks and columns] <br />
- atoms [columns] id, atom type, res number, res name, atom name, charge group, charge, mass (amu) <br />
- settles [columns] first atom, function (always 1), distance between O-H, distance between H-H (nm) <br />
- exclusions [columns] one atom index, atom indicies <br />
- bonds [columns] atom1, atom2, function, length(nm), force constant (kJ/mol/nm^2)<br />
- angles [columns] end atom, center atom, other end atom, equilib bond length (nm), force constant
