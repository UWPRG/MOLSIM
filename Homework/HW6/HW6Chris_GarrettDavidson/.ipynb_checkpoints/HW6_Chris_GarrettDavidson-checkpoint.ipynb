```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
data=np.loadtxt('VMD_dihedral_garrett.txt')
time=data[:,0]
dihedral=data[:,1]
new=[]
for angle in dihedral:
    new.append(math.radians(angle))
new_time=[]
for times in time:
    new_time.append(times/10)
plt.figure(figsize=(6, 6))
plt.plot(new_time,new,'r-')
plt.rc('font',size=12)
plt.title('Dihedral Angle vs. Time Calculated using VMD')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (radians)')
plt.xlim([0, 1000])
plt.ylim([-3.2,3.2])
plt.grid()

```

1. How does this differ from standard MetaD?
    
        Adding bias to already visited states but do not deposit bias around the transition state. Allows for analysis of transition state kinetics (since the transition state has been unaffected by bias) and allows for a lack of knowledge of the transition state. The lack of knowledge of the transition state also leads to a way of recognizing whether the transtion state has been crossed.
        
2. What are the advantages and disadvantages compared to standard MetaD?

    * Do not need knowledge of transition states
    * Useful in complex systems.
    * Keeps transition state intact so we can find a rate
