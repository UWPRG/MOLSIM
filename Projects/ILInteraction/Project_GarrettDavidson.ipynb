# Ionic Liquid Interaction Energies projects

Wes is trying to develop an equation of state model for ionic liquids to predict velocity. In order to do this, we need the interaction energies between these ions. 

I calculated the total energies of each ion separately in a vacuum, then with both ions in the same box. From there, I subtracted each of the individual ions from the salt to get the energies of the two ions interacting.

List of salts:
- BAM, FOR
- BMI, ACE
- BMI, TF2
- EMI, ACE
- EMI, ETS
- EMI, FMS
- EPY, ETS
- HMI, CHL
- HMI, TF2
- HPY, TF2
- OMI, CHL
- OMI, FMS
- OMI, TF2
- PAM, FOR
- PPY, TF2
- EAM, FOR
- BMI, MTS
- EMI, TF2
- EPY, TF2

Simulations of cation/anion systems in vacuum at 350K. Ran for 1ns.

```python
just_salt=[]
indexer=0
    

for salts in list_of_salts:
    for index in range(len(BAM_FOR)):
        just_salt.append(salts.salt.loc[index]-(salts.cation.loc[index]+salts.anion.loc[index]))
    
    average=np.average(just_salt)
    salt_array=np.array(just_salt)
    salt_array.transpose()
    salts['just_salt']=pd.DataFrame(salt_array)
    
    plt.figure(figsize=[7,5])
    #plt.subplot(10,2,indexer+1)
    plt.plot(salts['Step'],salts['cation'])
    plt.plot(salts['Step'],salts['anion'])
    plt.plot(salts['Step'],salts['salt'])
    plt.plot(salts['Step'],salts['just_salt'])
    plt.legend([list_of_salts_b[indexer][0:3:1],list_of_salts_b[indexer][4::],list_of_salts_b[indexer],'Subtracted'])
    plt.ylabel('Total Energy (kJ/mol)')
    plt.xlabel('Step')
    plt.title([list_of_salts_b[indexer],average])
    indexer+=1
print(indexer)
```

## Observations
Subtracted interaction energies are at the same value.  (-160 to -170 kJ/mol)

Salts with FOR anion always have lower interaction energies than the two individual molecules. 

Salts with TF2 or FMS anions always have higher interaction energies than the two individual molecules
